```python
import pandas as pd

from os.path import isfile
from tqdm.notebook import tqdm
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
def get_reranker_paragraphs(feature):
    paragraphs_path = f'/content/drive/MyDrive/Grammars Paragraphs/{feature}/Wikipedia 50/'
    reranker_paragraphs_path = f'/content/drive/MyDrive/Grammars Paragraphs/{feature}/Reranker 20/'
    reranker_scores_path = f'/content/drive/MyDrive/Reranker Similarity Scores/{feature}/scores.csv'
    df_reranker_scores = pd.read_csv(reranker_scores_path)
    grammar_files = df_reranker_scores.columns.values.tolist()
    for filename in tqdm(grammar_files):
        if not isfile(f'{reranker_paragraphs_path}{filename}.csv'):
            df = pd.read_csv(f'{paragraphs_path}{filename}.csv')
            df['Similarity Score'] = df_reranker_scores[filename]
            df = df.sort_values(by='Similarity Score', ascending=False, ignore_index=True)
            df = df.drop('Similarity Score', axis=1).head(20)
            df.to_csv(f'{reranker_paragraphs_path}{filename}.csv', index=False)
```

## WALS 81A: Order of Subject, Object and Verb

```python
get_reranker_paragraphs('WALS 81A')
```
